```python
def clean_brand(review):
    cleaned_review = review.strip()
    cleaned_review = re.sub(r'\r', '', cleaned_review)
    return cleaned_review

unique_brands = ['audi', 'bmw', 'buick', 'cadillac', 'chevrolet', 'chrysler',
        'dodge', 'ford', 'honda', 'hyundai', 'jaguar', 'jeep', 'kia',
        'lexus', 'mazda', 'mercedes‑benz', 'nissan', 'porsche', 'ram',
        'subaru', 'volvo', 'toyota', 'volkswagen']#, 'land-rover']

for brand in tqdm(unique_brands):
    print(brand, len(brand))
    vehicle_reviews = pd.read_csv(f'../data/Reviews/reviews_{str(brand)}.csv', lineterminator='\n', index_col='Unnamed: 0')
    vehicle_reviews['model_year'] = vehicle_reviews['Vehicle_Title'].apply(lambda x: x.split()[0] if x.split()[0].isdigit() else None)
    vehicle_reviews['model'] = vehicle_reviews['Vehicle_Title'].apply(lambda x: x.split()[2])
    vehicle_reviews['Brand'] = vehicle_reviews['Brand'].apply(lambda x: x.lower())
    vehicle_reviews['model_year'] = vehicle_reviews['model_year'].apply(lambda x: x.lower())
    vehicle_reviews['model'] = vehicle_reviews['model'].apply(lambda x: x.lower())
    vehicle_reviews['Review'] = vehicle_reviews['Review'].apply(lambda x: clean_brand(x))
    vehicle_reviews.to_csv(f'../data/Reviews/cleaned_reviews/reviews_{str(brand)}.csv', index=False)
    vehicle_reviews[['Brand', 'model']].drop_duplicates().to_csv(f'../data/Reviews/unique_reviews_by_brand/unique_reviews_{str(brand)}.csv', index=False)
vehicle_reviews
```

```python
def extract_info(text):
    pattern = r'(\d{4})\s(.+?)\s(.+)'
    match = re.match(pattern, text)
    if match:
        year = match.group(1)
        brand = match.group(2)
        model = match.group(3)
        return year, brand, model
    
    return None, None, None

msrp[['Year', 'Brand', 'Model']] = msrp['Make-Model'].apply(extract_info).apply(pd.Series)
msrp['Model'] = msrp.Model.apply(lambda x: re.sub(re.compile(r'\xa0'), ' ', x))
msrp['Brand'] = msrp['Brand'].apply(lambda x: x.lower())
msrp['Model'] = msrp['Model'].apply(lambda x: x.lower())
msrp[['Year', 'Brand', 'Model', 'MSRP Price', 'Starting Market Average']].to_csv('../data/MergeData_Test/Brand_Model_MSRP.csv', index=False)
```

## MSRP+Reviews+Specs & Safety

```python
nhtsa_df = pd.read_csv('../data/NHTSA.csv')
nhtsa_df = nhtsa_df.rename(columns={'Make': 'Brand'})
nhtsa_df['Brand'] = nhtsa_df['Brand'].apply(lambda x: x.lower())
nhtsa_df['Model'] = nhtsa_df['Model'].apply(lambda x: x.lower())
nhtsa_df['ModelYear'] = nhtsa_df['ModelYear'].apply(lambda x: int(x))
nhtsa_df = nhtsa_df.reset_index(drop=True)
vehicle_safety = nhtsa_df.dropna().reset_index(drop=True)
vehicle_safety.SafetyRatings = vehicle_safety.SafetyRatings.apply(lambda x: ast.literal_eval(x)[0])
vehicle_safety['Safety_Rating'] = vehicle_safety.SafetyRatings.apply(lambda x: x['OverallRating']) 
vehicle_safety['Complaints_Count'] = vehicle_safety.SafetyRatings.apply(lambda x: x['ComplaintsCount']) 
vehicle_safety = vehicle_safety.dropna()
vehicle_safety = vehicle_safety.loc[(vehicle_safety.Safety_Rating != 'Not Rated')]
vehicle_safety['Safety_Rating'] = vehicle_safety.Safety_Rating.apply(lambda x: int(x))
vehicle_safety['Complaints_Count'] = vehicle_safety.Complaints_Count.apply(lambda x: int(x))

# lets target certain models that can be easily corrected manually so they can easily be added to the db

#######################################
# audi
vehicle_safety.Model = vehicle_safety.Model.apply(lambda x: x.replace('audi ', '') if 'audi ' in x else x)
#######################################


vehicle_safety_grouped = vehicle_safety.groupby(['Brand', 'Model', 'ModelYear'])\
                        .agg(Avg_Safety_Rating=('Safety_Rating', 'mean'), 
                             Avg_Complaints_Count=('Complaints_Count', 'mean'))

vehicle_msrp_reviews_specs_safety = vehicle_msrp_reviews_specs.merge(vehicle_safety_grouped,
                                        how='inner',
                                        left_on = ['Brand', 'Model', 'Year'],
                                        right_on = ['Brand', 'Model', 'ModelYear'], copy=False).reset_index(drop=True)

vehicle_msrp_reviews_specs_safety
```
